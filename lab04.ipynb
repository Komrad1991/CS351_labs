Стоимость билета - хороший признак\
Проанализируйте этот параметр\
заменим билеты с нулевой стоимостью на медианную стоимость соотв класса


```python
med_price = {1:data.where(data.Pclass == 1).Fare.median(),
             2:data.where(data.Pclass == 2).Fare.median(),
             3:data.where(data.Pclass == 3).Fare.median()}
data.loc[data['Fare'] == 0,'Fare'] = data['Pclass'].map(med_price)
```

Проанализировать зависимость выживаемости от стоимости билета

```python
data.loc[data['Survived'] == 0].Fare.hist()
data.loc[data['Survived'] == 1].Fare.hist(alpha=0.5)
```

```python
sns.heatmap(data[['Survived','Fare']].corr(),annot=True)
```
